### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = (
    f"{url}api-key={nyt_api_key}&begin_date={begin_date}&end_date={end_date}"
    + f'&fq={filter_query}&sort={sort}&fl={field_list}'
)
#print(query_url)
```

```python
# Create an empty list to store the reviews
reviews_list = []

# loop through pages 0-19
page = 0

for page in range(0,20):
    # create query with a page number
    query_url_page = f"{query_url}&page={str(page)}"
    #print(query_url_page)
    #print(page)
      
    # Make a "GET" request and retrieve the JSON
    response = requests.get(query_url_page)
    response
    
    # Retrieve reviews
    reviews = response.json()

    # Print results in JSON format
    #print(json.dumps(reviews, indent=4))
    
    # Add a twelve second interval between queries to stay within API query limits

    #pd.Timedelta(seconds=12)
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
        
    try: 
        # loop through the reviews["response"]["docs"] and append each review to the list

        length_of_reviews_response_docs = len(reviews["response"]["docs"])
        print(f"Length of Reviews Docs: {length_of_reviews_response_docs}")

        for index in range(length_of_reviews_response_docs):
            #print(index)
            reviews_list.append(reviews["response"]["docs"][index])
           
        # Print the page that was just retrieved
        print(f"Current Page retrieved: {page}")
    
    except:
        # Print the page number that had no results then break from the loop
        print(f"Current Page retrieved had NO Results: {page}")
        break
        
        

     
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data

print(json.dumps(reviews_list[0:5], indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
reviews_list_df = pd.json_normalize(reviews_list)
reviews_list_df
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

#reviews_list_df.columns
#print(reviews_df["headline.main"])
reviews_list_df['title'] = reviews_list_df['headline.main'].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])
#print(reviews_list_df["title"])
reviews_list_df
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    #print(keyword_list)
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        #print(keyword)
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
#print(reviews_list_df["keywords"])
reviews_list_df["keywords"] = reviews_list_df["keywords"].apply(extract_keywords)
reviews_list_df.head()
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
movies_titles_list = reviews_list_df["title"].to_list()
movies_titles_list
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key

```

```python
# Create an empty list to store the results

tmdb_movies_list=[]


# Create a request counter to sleep the requests after a multiple
# of 50 requests

request_counter = 1

# Loop through the titles
#for title in movies_titles_list[197:200]: # only for debugging
for title in movies_titles_list:

    # Check if we need to sleep before making a request
    if request_counter % 50 == 0:
        time.sleep(1)
        print("It's time for a snooze.The application is sleeping.")

    # Add 1 to the request counter
    request_counter += 1
    
    # Perform a "GET" request for The Movie Database
    title_movie_url_query = url + title + tmdb_key_string
    #for debugging a single title: title_movie_url_query = url + movies_titles_list[0] + tmdb_key_string
    #print(title_movie_url_query)
    
    response = requests.get(title_movie_url_query)
    response
    
    # Retrieve reviews
    movie_reviews = response.json()

    # Print results in JSON format
    #print(json.dumps(movie_reviews, indent=4))

    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    
    try:
        # Get movie id
        movie_id = movie_reviews["results"][0]['id']   
        #print(movie_id) 

        movie_id_url = f"https://api.themoviedb.org/3/movie/{movie_id}?{tmdb_key_string}"
        #print(movie_id_url)
    
        # Make a request for a the full movie details
        # Execute "GET" request with url

        response = requests.get(movie_id_url)
        response   
    
        # Retrieve reviews
        movie_id_reviews = response.json()

        # Print results in JSON format
        #print(json.dumps(movie_id_reviews, indent=4))
    
        genres_names_list = []
        spoken_languages_english_list = []
        production_countries_names_list = []
    
        # Extract the genre names into a list
        
        genre_names_list = [genre['name'] for genre in movie_id_reviews['genres']]
        #print(genre_names_list)
        
        # Extract the spoken_languages' English name into a list
        
        spoken_languages_english_list = [spoken['english_name'] for spoken in movie_id_reviews['spoken_languages']]
        #print(spoken_languages_english_list)
        
        # Extract the production_countries' name into a list

        production_countries_names_list = [production['name'] for production in movie_id_reviews['production_countries']]
        #print(production_countries_names_list)
    
        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        
        #print(f"Movie_id_reviews Title {movie_id_reviews['title']}") for debugging
        #print(f"Just Title {title}") for debugging
        
              
        tmdb_movies_dict = {"title":movie_id_reviews['title'],
                            "original_title":movie_id_reviews['original_title'],
                            "budget": movie_id_reviews['budget'],
                            "genre": genre_names_list,
                            "language": movie_id_reviews['original_language'],
                            "spoken_languages": spoken_languages_english_list,
                            "homepage": movie_id_reviews['homepage'],
                            "overview": movie_id_reviews['overview'],
                            "popularity": movie_id_reviews['popularity'],
                            "runtime": movie_id_reviews['runtime'],
                            "revenue": movie_id_reviews['revenue'],
                            "release_date": movie_id_reviews['release_date'],
                            "vote_average": movie_id_reviews['vote_average'],
                            "vote_count": movie_id_reviews['vote_count'],
                            "production_countries": production_countries_names_list
                           }
        #print(tmdb_movies_dict)
    
        tmdb_movies_list.append(tmdb_movies_dict)
        #print(tmdb_movies_list)
        
        # Print out the title that was found
        print(f"Found Movie: '{title}' for movie_id {movie_id}")
 
              
    except:
        # Print the movie_id and title that had no results 
        print(f"No Results for Movie: '{title}', Movie not found")
                
        


        


        

       
        
       
```

```python
# Convert the results to a DataFrame
tmdb_df = pd.DataFrame(tmdb_movies_list)
tmdb_df
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merged_nytr_tmdbdf = pd.merge(tmdb_df, reviews_list_df, how="inner", on="title")
merged_nytr_tmdbdf
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing

columns_to_fix = ['genre','spoken_languages','production_countries']
# Create a list of characters to remove
characters_to_remove = ["[","]","'"]

# Loop through the list of columns to fix
for col_tofix in columns_to_fix:

    # Convert the column to type 'str'
    merged_nytr_tmdbdf[col_tofix] = merged_nytr_tmdbdf[col_tofix].astype('str')

    # Loop through characters to remove
    for chars_toremove in characters_to_remove:
        merged_nytr_tmdbdf[col_tofix] = merged_nytr_tmdbdf[col_tofix].str.replace(chars_toremove,"",regex=False)


# Display the fixed DataFrame
#merged_nytr_tmdbdf['genres']
#merged_nytr_tmdbdf['spoken_languages']
#merged_nytr_tmdbdf['production_countries']
merged_nytr_tmdbdf
```

```python
# Drop "byline.person" column
drop_byline_nytr_tmdbdf= merged_nytr_tmdbdf.drop('byline.person',axis=1)
drop_byline_nytr_tmdbdf
```

```python
# Delete duplicate rows and reset index
drop_nytr_tmdbdf = drop_byline_nytr_tmdbdf.drop_duplicates(keep='first').reset_index(drop=True)
#drop_nytr_tmdbdf = drop_byline_nytr_tmdbdf.reset_index(drop=True)
drop_nytr_tmdbdf
```

```python
# Export data to CSV without the index
drop_nytr_tmdbdf.to_csv("final_nytr_tmdb.csv",index=False,header=True)
```
